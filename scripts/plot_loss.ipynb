```python
import numpy as np
import matplotlib.pyplot as plt
import mplhep as hep

plt.style.use(hep.cms.style.ROOT)
```

```python
models   = ["fgsm/"]
adv      = [True]
```

```python
def plot_loss(model_names, is_adv):
    base_input_dir  = "/hpcwork/pj214607/work/promotion/deepjet/results/"
    base_output_dir = "/home/pj214607/work/promotion/deepjet/results/"
    
    for i,name in enumerate(model_names):
        if is_adv[i]:
            num_epochs = np.arange(1,79)
        else:
            num_epochs = np.arange(1,40)
            
        loss = np.load(base_input_dir + name + "loss.npy")
        
        plt.figure()
        plt.plot(num_epochs, loss[0], label="train loss", color="blue")
        plt.plot(num_epochs, loss[1], label="val loss", color="orange")
        plt.legend(loc="best")
        plt.grid(which="minor", alpha=0.85)
        plt.grid(which="major", alpha=0.95, color='black')
        plt.xlabel("epoch")
        plt.ylabel("loss")
        plt.title(name[:-1] + " training")
        plt.savefig(base_output_dir + name + "loss.pdf")
        plt.show()
        plt.close()
```

```python
plot_loss(models, adv)
```
